```python
ax = sns.histplot(x='loan_amount_requested', data=df, kde=True, color='purple',)
plt.axvline(df['loan_amount_requested'].mean(), color ='blue', linestyle = '--', lw = 1, label='Mean')
plt.axvline(df['loan_amount_requested'].median(), color = 'green', linestyle = '-', lw = 1, label = 'Median')
```

```python
# The bell curve confirms the data is positively skewed. The smaller the loan amounts requested, the higher the count. The higher the amounts requested the lower the count.  
```

```python
average_age_income = df.groupby('applicant_age')['monthly_income'].mean().reset_index()
```

```python
plt.figure(figsize=(12,4))
sns.lineplot(x='applicant_age', y='monthly_income', data = average_age_income)
plt.title('Average Income by Age')
```

```python
# This suggests that there is no strong linear relationship between an applicants age and monthly income
    # Age is not a reliable indicator for monthly income 
        #Between ages 50â€“60, income fluctuates the most, which could mean:
            #Some people are at peak career earnings
            #Others are retired, semi-retired, or in lower-paid roles
```
